```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
file_path = 'USA_cars_datasets.csv'
car_data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
car_data.head()
```

```python
# Check for missing values and data types
missing_values = car_data.isnull().sum()
data_types = car_data.dtypes
print("Missing Values:\n", missing_values)
print("\nData Types:\n", data_types)
```

```python
#Price Trends Analysis
# Average Car Price by Year
average_price_by_year = car_data.groupby('year')['price'].mean().reset_index()
```

```python
# Plotting the average car price by year
plt.figure(figsize=(10, 6))
plt.plot(average_price_by_year['year'], average_price_by_year['price'], marker='o', linestyle='-')
plt.title('Average Car Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Price (USD)')
plt.grid(True)
plt.show()
```

The chart clearly indicates that the average car prices have been rising significantly in recent years. The steep decline in the earlier years might be due to data limitations or specific market conditions at that time. The consistent rise in prices from the early 2000s to 2020 highlights a trend of increasing car values, reflecting broader market trends, technological advancements, and economic factors. This information can be useful for potential buyers and sellers to understand historical price trends and make informed decisions in the car market.

```python
# Average Car Price by Brand
average_price_by_brand = car_data.groupby('brand')['price'].mean().sort_values(ascending=False).reset_index()

# Plotting the average car price by brand
plt.figure(figsize=(12, 8))
plt.barh(average_price_by_brand['brand'], average_price_by_brand['price'], color='skyblue')
plt.title('Average Car Price by Brand')
plt.xlabel('Average Price (USD)')
plt.ylabel('Brand')
plt.gca().invert_yaxis()
plt.show()
```

# The chart illustrates a clear segmentation of car brands based on their average prices:
Luxury and High-End Brands: Harley-Davidson, Lexus, Maserati, Land Rover, Mercedes-Benz, and BMW lead the market with the highest average prices, reflecting their premium status.
Mid-Range Brands: Cadillac, Ford, Buick, Chevrolet, Lincoln, and Dodge offer a balance of luxury and affordability.
Affordable Brands: Nissan, Kia, Ram, Jeep, GMC, Mazda, Acura, Toyota, Honda, and Hyundai cater to the budget-conscious consumers, offering reliable and economical options.
Lowest Priced Brands: Heartland, Jaguar, and Peterbilt have the lowest average prices, potentially due to niche market focus or specific models.

```python
# Mileage Trends Analysis

# Scatter plot of car price vs. mileage
plt.figure(figsize=(10, 6))
plt.scatter(car_data['mileage'], car_data['price'], alpha=0.5, color='blue')
plt.title('Car Price vs. Mileage')
plt.xlabel('Mileage')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

The scatter plot effectively highlights the negative correlation between car mileage and price:

Lower Mileage, Higher Prices: Cars with lower mileage are priced higher, reflecting their better condition and less wear.
Higher Mileage, Lower Prices: As mileage increases, car prices decrease significantly, indicating reduced value due to higher usage.
Market Insights: Buyers looking for budget-friendly options might consider higher-mileage cars, while sellers with low-mileage vehicles can expect higher prices.

```python
#State-Wise Analysis

# Average Car Price and Mileage by State
average_price_mileage_by_state = car_data.groupby('state').agg({'price': 'mean', 'mileage': 'mean'}).reset_index()

# Plotting the average car price by state
plt.figure(figsize=(12, 8))
plt.bar(average_price_mileage_by_state['state'], average_price_mileage_by_state['price'], color='green')
plt.title('Average Car Price by State')
plt.xlabel('State')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

The bar chart highlights significant state-wise differences in average car prices:

Kentucky has the highest average car prices, which is a notable outlier compared to other states.
Several states, including California, Texas, and New York, show moderate car prices, reflecting their diverse markets with both economical and premium options.
States with lower average car prices, such as Alabama and West Virginia, might have more affordable car options or a preference for used vehicles.

```python
# Plotting the average car mileage by state
plt.figure(figsize=(12, 8))
plt.bar(average_price_mileage_by_state['state'], average_price_mileage_by_state['mileage'], color='orange')
plt.title('Average Car Mileage by State')
plt.xlabel('State')
plt.ylabel('Average Mileage')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

Montana stands out with the highest average car mileage among all states. This suggests that cars in Montana tend to be driven more extensively than those in other states, possibly due to the larger geographical area and rural nature of the state, leading to longer travel distances.

```python
# Brand Popularity

# Count the number of listings for each brand
brand_popularity = car_data['brand'].value_counts().reset_index()
brand_popularity.columns = ['brand', 'number_of_listings']

# Plotting the brand popularity by number of listings
plt.figure(figsize=(12, 8))
plt.bar(brand_popularity['brand'], brand_popularity['number_of_listings'], color='purple')
plt.title('Brand Popularity by Number of Listings')
plt.xlabel('Brand')
plt.ylabel('Number of Listings')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

The bar chart reveals significant insights into brand popularity in the used car market:

Ford is the most popular brand with the highest number of listings, indicating its widespread availability and preference among buyers.
Other brands like Dodge, Nissan, Chevrolet, and GMC also show strong market presence.
Less popular brands still maintain a niche market but have fewer listings.
This information can help buyers understand the market availability of different brands and sellers to recognize the competitive landscape.



Through our exploration, we uncovered several key insights:
1. **Price Trends**: Car prices have generally increased over the years, as seen in the 'Average Car Price by Year' plot.
2. **Brand Analysis**: Certain brands, such as Chevrolet and Ford, command higher average prices compared to others, and their popularity varies widely, as shown in the 'Average Car Price by Brand' and 'Brand Popularity by Number of Listings' charts.
3. **Mileage Impact**: There is a clear negative correlation between car mileage and price, indicating that cars with lower mileage tend to have higher prices, as illustrated in the 'Car Price vs. Mileage' scatter plot.
4. **State Variations**: Car prices and mileage exhibit regional differences, indicating varying market conditions across states, as demonstrated in the 'Average Car Price by State' and 'Average Car Mileage by State' bar charts.
5. **Brand Popularity**: Some brands are more frequently listed than others, reflecting their market presence, as highlighted in the 'Brand Popularity by Number of Listings' chart.

By weaving these insights into a cohesive story, we can provide a comprehensive overview of the used car market in the USA. These findings can help potential buyers and sellers make informed decisions based on price trends, brand value, mileage considerations, and regional differences.
